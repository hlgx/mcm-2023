```python
def max_repeating_char(string):
    char_counts = {}
    for char in string:
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1
    max_count = 0
    max_char = None
    for char, count in char_counts.items():
        if count > max_count:
            max_count = count
            max_char = char
    return max_count-1
```

```python
def get_all_letters(str):  
    multi = 1
    for char in str:
        if char == 'a':
            multi *= 8.47
        if char == 'b':
            multi *= 2.43
        if char == 'c':
            multi *= 4.13
        if char == 'd':
            multi *= 3.40
        if char == 'e':
            multi *= 10.67
        if char == 'f':
            multi *= 1.99
        if char == 'g':
            multi *= 2.70
        if char == 'h':
            multi *= 3.37
        if char == 'i':
            multi *= 5.80
        if char == 'j':
            multi *= 0.23
        if char == 'k':
            multi *= 1.82
        if char == 'l':
            multi *= 6.22
        if char == 'm':
            multi *= 2.73
        if char == 'n':
            multi *= 4.97
        if char == 'o':
            multi *= 6.53
        if char == 'p':
            multi *= 3.17
        if char == 'q':
            multi *= 0.25
        if char == 'r':
            multi *= 7.78
        if char == 's':
            multi *= 5.79
        if char == 't':
            multi *= 6.31
        if char == 'u':
            multi *= 4.04
        if char == 'v':
            multi *= 1.32
        if char == 'w':
            multi *= 1.69
        if char == 'x':
            multi *= 0.32
        if char == 'y':
            multi *= 3.68
        if char == 'z':
            multi *= 0.35
    return multi
```
